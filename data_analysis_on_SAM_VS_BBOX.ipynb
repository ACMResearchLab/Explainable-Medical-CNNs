```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open("/content/drive/MyDrive/dataresults.json", "r") as read_file:
    data = json.load(read_file)
```

```python
def plot_distribution(data, title):
    plt.hist(data, bins=10, edgecolor='black')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.show()
```

```python
def KDE(data1, data2, label1, label2):
    sns.kdeplot(data1, label=label1)
    sns.kdeplot(data2, label=label2)
    plt.xlabel('Values')
    plt.ylabel('Density')
    plt.title('Distribution Comparison')
    plt.legend()
    plt.show()
```

```python
def plot_multiple_distribution(data1, data2, label1, label2):
    plt.hist(data1, bins=10, alpha=0.5, label=label1, edgecolor='black')
    plt.hist(data2, bins=10, alpha=0.5, label=label2, edgecolor='black')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Distribution Comparison')
    plt.legend()
    plt.show()
```

```python
clean = {"SAM AUPRC":[], "Bounded Box AUPRC":[], "SAM AUC Judd":[], "Bounded Box AUC Judd":[]}

for key, value in data.items():
  clean["SAM AUPRC"].append(value[0]["auprc"])
  clean["Bounded Box AUPRC"].append(value[1]["auprc"])
  clean["SAM AUC Judd"].append(value[0]["auc_judd"])
  clean["Bounded Box AUC Judd"].append(value[1]["auc_judd"])



print("TOTAL NUM: " + str(len(clean["SAM AUPRC"])))
print(statistics.mean(clean["SAM AUPRC"]))
print(statistics.mean(clean["Bounded Box AUPRC"]))
print()
print(statistics.mean(clean["SAM AUC Judd"]))
print(statistics.mean(clean["Bounded Box AUC Judd"]))
```

```python
print(statistics.stdev(clean["SAM AUPRC"]))
print(statistics.stdev(clean["Bounded Box AUPRC"]))
print()
print(statistics.stdev(clean["SAM AUC Judd"]))
print(statistics.stdev(clean["Bounded Box AUC Judd"]))
```

SAM auprc

```python
for keys, values in clean.items():
  plot_distribution(values, keys)
  print()
```

Distribution comparisons

---



```python
KDE(clean["SAM AUPRC"], clean["Bounded Box AUPRC"], "SAM AUPRC", "Bounding Box AUPRC")
plot_multiple_distribution(clean["SAM AUPRC"], clean["Bounded Box AUPRC"], "SAM AUPRC", "Bounding Box AUPRC")
```

```python
KDE(clean["SAM AUC Judd"], clean["Bounded Box AUC Judd"], "SAM AUC Judd", "Bounding Box AUC Judd")
plot_multiple_distribution(clean["SAM AUC Judd"], clean["Bounded Box AUC Judd"], "SAM AUC Judd", "Bounding Box AUC Judd")
```
